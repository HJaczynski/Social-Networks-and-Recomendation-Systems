## https://networkrepository.com/inf.php

```python
import numpy as np
import networkx as nx
import matplotlib.pyplot as plt
from matplotlib.colors import Normalize

G = nx.read_edgelist("networks/inf-euroroad/inf-euroroad.edges",
                     create_using=nx.Graph(), nodetype=int, comments="%")



pos = nx.kamada_kawai_layout(G)

deg = dict(G.degree())
dvals = np.array([deg[n] for n in G.nodes()])
sizes = 6 + 2.5*np.sqrt(dvals)

fig, ax = plt.subplots(figsize=(12, 12))

nx.draw_networkx_edges(G, pos, ax=ax, width=1, alpha=0.3)

nodes = nx.draw_networkx_nodes(
    G, pos, ax=ax,
    node_size=sizes,
    node_color=dvals,
    cmap=plt.cm.viridis,
    linewidths=0
)

norm = Normalize(vmin=dvals.min(), vmax=dvals.max())
cbar = fig.colorbar(plt.cm.ScalarMappable(norm=norm, cmap=plt.cm.viridis),
                    ax=ax, fraction=0.03, pad=0.01)
cbar.set_label("Node degree")

ax.set_axis_off()
plt.tight_layout()
plt.show()
```

```python
import scipy.io, scipy.sparse as sp, networkx as nx

M = scipy.io.mmread("networks/power-1138-bus/power-1138-bus.mtx").tocsr()

G = nx.from_scipy_sparse_array(M)

n_nodes = G.number_of_nodes()
n_loops = nx.number_of_selfloops(G)
loop_nodes = set(nx.nodes_with_selfloops(G))
frac = len(loop_nodes) / n_nodes

print(f"nodes: {n_nodes}")
print(f"self-loop edges: {n_loops}")
print(f"nodes with a self-loop: {len(loop_nodes)} ({frac:.1%})")
```

## https://networkrepository.com/rt.php

```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.colors import Normalize
import networkx as nx
import pandas as pd

df = pd.read_csv(
    "networks/rt_obama/rt_obama.edges",
    header=None, names=["src", "dst", "timestamp"]
)

G = nx.from_pandas_edgelist(
    df, source="src", target="dst", edge_attr=["timestamp"], create_using=nx.Graph()
)

print("Nodes:", G.number_of_nodes(), "Edges:", G.number_of_edges())

pos = nx.spring_layout(G)

deg = dict(G.degree())
dvals = np.array([deg[n] for n in G.nodes()])

sizes = 6 + 3 * np.sqrt(dvals)
norm = Normalize(vmin=dvals.min(), vmax=dvals.max())
cmap = plt.cm.viridis

fig, ax = plt.subplots(figsize=(12, 12))

nx.draw_networkx_edges(G, pos, ax=ax, width=1, alpha=0.3)

nodes = nx.draw_networkx_nodes(
    G, pos, ax=ax,
    node_size=sizes,
    node_color=dvals,
    cmap=cmap,
    linewidths=0
)

cbar = fig.colorbar(
    plt.cm.ScalarMappable(norm=norm, cmap=cmap),
    ax=ax, orientation="vertical", fraction=0.03, pad=0.01
)
cbar.set_label("Node Degree")

ax.set_axis_off()
plt.tight_layout()
plt.show()

```
