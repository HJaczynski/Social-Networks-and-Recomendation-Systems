```python
import networkx as nx 
import matplotlib.pyplot as plt

G = nx.read_edgelist(
    "networks/inf-euroroad/inf-euroroad.edges", 
    create_using=nx.Graph(), 
    nodetype=int, 
    comments="%"
)

deg = dict(G.degree())
sizes = [1 + 2*deg[n]**0.7 for n in G.nodes()] 

pos = nx.spring_layout(G)
plt.figure(figsize=(12, 12))
nx.draw_networkx_edges(G, pos, alpha=0.1, width=1)
nx.draw_networkx_nodes(G, pos, node_size=sizes, cmap="tab20")
plt.show()
```
